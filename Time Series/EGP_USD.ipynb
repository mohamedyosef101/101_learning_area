# Egyptian Pound vs US Dollar
**Data source**: [USD EGP Historical Data](https://www.investing.com/currencies/usd-egp-historical-data). Investing.com

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# styling
# set the visualization style
sns.set_style('whitegrid', {'grid.color': 'lightgray'})

# read the data
URL = 'https://github.com/mohamedyosef101/101_learning_area/raw/area/Time%20Series/data/USD_EGP.csv'
data = pd.read_csv(URL)
data.head()
```

```python
data = data.rename(columns = {'Date': 'date', 'Price': 'price',
                              'Open': 'open', 'High': 'high',
                              'Low': 'low', 'Change': 'change'})
data = data.set_index('date')
data.index = pd.to_datetime(data.index)
data = data.sort_index(ascending=True)
data = data.drop('Vol', axis=1)
data.head(3)
```

```python
plt.plot(data.price, color='lightcoral')
```
