# Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I do love Data Science and Research")
```

# Zero-shot classification

```python
from transformers import pipeline as p
c = p("zero-shot-classification")
c(
    "I have a project on Finance Services",
    candidate_labels=["education", "business", "sport"]
)
```

# Text generation

```python
text_gen = p("text-generation")
text_gen("In our life, we need to")
```

# Using Models

```python
generator = p("text-generation", model="distilgpt2")
generator(
    "In our life, we need to",
    max_length=42,
    num_return_sequences=2,
)
```

# Mask filling

```python
unmasker = p("fill-mask")
unmasker("Life will teach you <mask> lessons.", top_k=1)
```

# Named entity recognition

```python
NER = p("ner", grouped_entities=True)
NER("My name is Mohamed Yosef and I work as a data scientist in Egypt")
```

# Summarization

```python
summary = p("summarization")
summary(
    """
    Model-free policy control
Model-free policy control refers to methods of learning policies in
reinforcement learning without explicitly modeling the environment dynamics.
(Yes. I just want you to understand what model-free means)

Two main approaches are on-policy and off-policy control.

Approach 1. On-policy control
On-policy control refers to learning about a policy using experiential data
generated by following that same policy. The agent simultaneously evaluates and
improves the policy that is actively being executed. This approach has the
benefit of naturally learning about and optimizing the agent’s current way of
behaving. However, it requires the agent to explore new actions to improve.

On-policy -> following the same policy

Approach 2. Off-policy control
Off-policy methods allow the agent to learn about and evaluate a particular
target policy independently from the behavioral policy being followed. This
decouples the agent’s actions for exploration from the policy being optimized.
Algorithms like Q-learning are off-policy, learning about the greedy policy
while experimenting more randomly. Off-policy control is sample efficient
since all experience can inform the target policy.

Off-policy -> following a different policy
    """
)
```
